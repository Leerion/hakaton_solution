```python
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
```

```python
df = pd.read_excel('dataset_v2.xlsx')
```

```python
df_res = pd.DataFrame(columns=['Название', 'Ареал'])
df_res
```

```python
desc = df['Ареал'].values
ec = df['Экология'].values
names = df['name'].values
```

```python
for i in range(len(desc)):
    print(i)
    name = names[i]
    
    if not (pd.isna(desc[i])):
        s = desc[i]
    else:
        s = ec[i]
    
    text_file = open("inputPlants.txt", "w")
    text_file.write(s)
    text_file.close()
    !./tomita-mac config.proto

    with open("pretty.html", "r") as f:
        contents = f.read()

    soup = BeautifulSoup(contents, 'xml')

    for table in soup.find_all("tbody"):
        areals = []
        if (table.find_all("a")):
            for i in range(len(table.find_all("a"))):
                elem = table.find_all("a")[i].text
                areals.append(elem)
        areals = list(set(areals))
        dct = {'Название':name, 'Ареал':areals}
        df_res = df_res.append(dct, ignore_index = True)
```
