```python
import pandas as pd
from bs4 import BeautifulSoup
from pymystem3 import Mystem
import numpy as np
```

```python
#pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_excel('./dataset_v2.xlsx')
```

```python
df_batch = df
```

```python
desc = df_batch['Описание'].values
names = df_batch['name'].values
soils = df_batch['Экология'].values
text = [x+y for x, y in zip(desc, soils)]
```

```python
df_res = pd.DataFrame(columns=['Название', 'Период сбора урожая', 'Типы почв'])

for i in range(len(desc)):
    print(i)
    s = text[i]
    name = names[i]
    
    
    text_file = open("inputPlants.txt", "w")
    text_file.write(s)
    text_file.close()
    !./tomita-mac config.proto
    
    with open("pretty.html", "r") as f:
        contents = f.read()
        
    soup = BeautifulSoup(contents, 'xml')
    
    month = np.NaN
    soil_type = np.NaN
    for table in soup.find_all("tbody"):
        if (table.find_all("b")):
            col = table.find_all("b")[0].text
            if col=="Harvest":
                elem = table.find_all("a")[0].text
                month = elem

            elif col=="Soil":
                elem = []
                for i in range(len(table.find_all("a"))):
                    elem.append(table.find_all("a")[i].text)
                soil_type = elem

    dct = {'Название':name, 'Период сбора урожая':month, 'Типы почв':soil_type}
    df_res = df_res.append(dct, ignore_index = True)
```

```python
def get_month(t):
    m = Mystem()
    t_copy = t
    if not pd.isna(t):
        lemmas = m.lemmatize(t)

        t = ''.join(lemmas)
        t = t.replace('\n', '').replace(' .', '')
        t = t.split(' ')
        t = t[-1]
        t = t.split('-')

        months = ['январь', 'февраль', 'март', 'апрель', 'май', 'июнь', 'июль', 'август', 'сентябрь', 'октябрь', 'ноябрь', 'декабрь']
        if t[0] in months:
            start = months.index(t[0])
            end = months.index(t[-1])
            month = months[start:end+1]
            return month
    return np.NaN
```

```python
df_res['Месяц']= df_res['Период сбора урожая'].apply(lambda x: get_month(x))
```
